1 - DEF - função nomeada

```python
# Criação de uma função de soma, com dois parâmetros
def somar_dois_valores(primeiro, segundo):
  return primeiro + segundo
```

```python
# Chamando a função e passando dois argumentos solicitados
somar_dois_valores(10, 15)
```

```python
# Criando uma função de transformação do texto em letras maiúsculas, com um parâmetro
def texto_maiusculo(texto):
  return texto.upper()
```

```python
# Chamando a função e passando o argumento solicitado
texto_maiusculo("live de python show é com trovato e wallace carlis")
```

2 - LAMBDA - função anônima

```python
# Criando uma função lambda de soma de dois parâmetros
lambda x, y: x + y (10, 15)
```

```python
# Criando uma função lambda de soma de dois parâmetros com parênteses
(lambda x, y: x + y) (10, 15)
```

```python
# Criando uma função lambda com atribuição de variável
soma = lambda x, y: x + y
soma(10, 15)
```

3 - FOR - instrução / estrutura de controle

```python
# Criando um For com um argumento:
for i in range(3):
  print(i)
```

```python
# Criando um For com dois argumentos:
for i in range(0, 3):
  print(i)
```

```python
# Criando um For com índice 1:
for i in range(1, 3):
  print(i)
```

```python
# Criando um For com dois argumentos e soma na iteração:
for i in range(0, 3):
  print(i + 10)
```

11 - ASTYPE() - função de transformação do tipo de dado

```python
# Visualizando a base df_str
df_str
```

```python
# Transformando o tipo de dado da coluna idade da base df_str para int com a função astype()
df_str.idade = df_str.idade.astype(int)
df_str
```

```python
# Verificando o valor da média da idade de df_str após a transformação
df_str.idade.mean().round()
```

```python
# Verificando novamente as principais informações da base df_str com a função info()
df_str.info()
```

12 - LIST COMPREHENSION [ ] - lista com um for interno

```python
# Visualizando a base df_nan
df_nan
```

```python
# Tentando transformar a coluna idade da base df_float em int com List Comprehension - erro de valor
df_nan["idade"] = [int(x) for x in df_nan.idade]
```

```python
# Obtendo a média da coluna idade
media_idade = df_nan.idade.mean().round()
media_idade
```

13 - FILLNA() - preenchimento de valores NaN

```python
# Transformando a coluna idade da base df_nan em int com List Comprehension e a função fillna()
df_nan.idade = [int(x) for x in df_nan.idade.fillna(value = media_idade)]
df_nan
```

14 - APPLY LAMBDA - aplicação de funções em conjunto

```python
# Visualizando a base de dados df_float
df_float
```

```python
# Transformando a coluna de idade da base df_float com as funções apply() e lambda
df_float["idade"] = df_float["idade"].apply(lambda x: int(x))
df_float
```

15 - LIST MAP LAMBDA - mapeamento e transformação com três funções

```python
# Visualizando a base de dados df
df
```

```python
# Criando uma coluna de salário com idade multiplicada por 5000 com as funções list(), map() e lambda
df["salario"] = list(map(lambda x: x * 5000, df.idade))
df
```
